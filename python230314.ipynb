## 변수의 종류 (Scala 자료형)
- 크게 4가지 존재 : int, float, bool, None
- 더 이상 나누어지지 않는 것

### int 형 예제

```python
temp_int = 1
print(temp_int)
print(type(temp_int))
```

- 문자열

```python
temp_int = '1'
print(temp_int)
print(type(temp_int))
```

### float형 예제
- 실수형

```python
num_float = 0.2
print(num_float)
print(type(num_float))
```

### Bool 형

```python
temp_bool = True
temp_bool = False
print(temp_bool)
print(type(temp_bool))
```

## 사칙연산
- 정수형 사칙연산

```python
a=6.0
b=2.0

print(a+b)
print(a-b)
print(a*b)
print(a/b) #나누기 할시 실수형으로 떨어짐,(딱맞아 떨어져도 .0이 표시되어 실수형)
print(a//b) #몫
print(a%b)  #나머지
print(a**b) #제곱
```

### 논리형 연산자
- AND 조건, OR 조건이 있음

```python
print(True and True)
print(True and False)
print(False and True)
print(False and False)
print("----------")
print(True or True)
print(True or False)
print(False or True)
print(False or False)
```

### 비교 연산자
- 비교연산자는 부등호를 의미함
- 왼쪽이 기준 값

```python
print(4>3)
print(3>4)
print(3>=4)
print(3<=4)
```

## input()
- print() : 출력
- input() : 입력

```python
var = int(input("값을 입력해주세요!!"))
print(var)
print(type(var))
```

```python
var = input("값을 입력해주세요!!")
print(var)
var = int(var)
print(type(var))
```

```python
num1 = input("값을 입력해주세요!! ")
num2 = input("값을 입력해주세요!! ")

result = int(num1)+int(num2)
print(result)
```
